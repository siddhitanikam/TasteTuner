```python
import pandas as pd
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Load the CSV file
df = pd.read_csv('Ingredient Data.csv')  # Replace 'path_to_your_file.csv' with your actual file path

# Combine all the ingredient names into a single string
text = ' '.join(ingredient for ingredient in df['name'].dropna())  # Make sure there are no NaN values

# Create the word cloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)

# Display the word cloud using matplotlib
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')  # Hide the axes
plt.show()
```
